# SNR 169 (IC443)

```python
xr, yr, image_data =  radioar.readfits('snr169')
xgrid, ygrid, ar, realdis = radioar.readsnrsav('snr169')
```

```python
radioar.radio_ar('snr169', ar, realdis, image_data, xr, yr, xgrid, ygrid, perpc=True, single = False,nradius=2, vmax=1500 )
```

```python
radioar.radio_ar('snr169', ar, realdis, image_data, xr, yr, xgrid, ygrid, perpc=True, single = True, nradius=2, vmax=1500, picn=4, cloudn=1)
```

```python
for i in range(1):
    dis_ext.datainborder('snr169', cloudn = i+1)
    pfit, perror = dis_ext.disarMCMCall('snr169', strategy='MC', cloudn = i+1)
    print('Fitted paras for cloud {0}: '.format(i+1), pfit)
    print('Errors for cloud {0}: '.format(i+1), perror)
```

```python
for i in range(1):
    dis_ext.datainborder('snr169', cloudn = i+1)
    pfit, perror = dis_ext.disarMCMCall('snr169', strategy='MC', cloudn = i+1, nsigma_down = 3, nsigma_up = 3)
    print('Fitted paras for cloud {0}: '.format(i+1), pfit)
    print('Errors for cloud {0}: '.format(i+1), perror)
```
