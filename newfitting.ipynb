```python
import SNR13
import _pickle
import numpy as np
import matplotlib.pyplot as plt
from os.path import join as pjoin
import matplotlib
import matplotlib.ticker as ticker
```

```python
from SNR13 import dis_ext
```

```python
def binsigma(data, start, end, length):
    '''std of samples in each bin
    Args:
        data: 2-d array, [dis, ar]
        start, end, length: imformation for setup bins
    Returns:
        dismean: 1-d array of means of distance data
        arsigma: 1-d array of std of samples in each bin
    '''
    #
    data=np.array(data)
    bin_number = int((end - start) // length)  #decide how many bins will be choosen
    arlist = []
    dislist = []
    for i in np.arange(bin_number):
        index = np.logical_and(data[0] > start + i * length, data[0] < start + (i + 1) * length)
        if np.count_nonzero(index) >= 1:
            binar = data[1][index]
            bindis = data[0][index]
            binstd = np.std(binar)
            binmean = np.mean(bindis)
        else:
            binstd = []
            binmean = []
        arlist.append(binstd)
        dislist.append(binmean)
    arsigma = np.array(arlist)  #join a sequence of arrays along an existing axis
    dismean = np.array(dislist)
    return dismean, arsigma
```

```python
name = 'snr170'
cloudn = 2
datapath = pjoin('..', '..', 'Data', 'extin3d', 'results', '{0}_diar_{1}.pkl'.format(name, cloudn))
datafile = open(datapath, 'rb')
data = _pickle.load(datafile)
dis, ar = data.dis, data.ar
datafile.close()

disx, arx = dis_ext.sigmaclips(np.array([dis,ar]), start=0.3, end=3.0, length=0.1, nsigma_down=2, nsigma_up=2)
dismean, arsigma = binsigma(np.array([disx, arx]), start=0.3, end=3.0, length=0.1)

matplotlib.rcdefaults()
p = matplotlib.rcParams
p["figure.subplot.left"] = 0.125
p["figure.subplot.right"] = 0.95   
p["figure.subplot.bottom"] = 0.12  
p["figure.subplot.top"] = 0.95   
p["figure.subplot.wspace"] = 0.05
p["figure.subplot.hspace"] = 0.05
fig = plt.figure(figsize=(4.5,4))
plt.rc('text', usetex=True)
ax = fig.add_subplot(111)
#stars = ax.plot(dis, ar, '.k', ms = 3, label = r'$\rm Sources~in~selected~area$')
starsx = ax.plot(dismean, arsigma, 'g', ms = 5)
ax.xaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(0.1))
ax.yaxis.set_major_locator(ticker.MultipleLocator(0.1))
ax.yaxis.set_minor_locator(ticker.MultipleLocator(0.01))
ax.tick_params(labelsize=12)
xla = ax.set_xlabel(r"$\rm d~(kpc)$", fontsize=12)
yla = ax.set_ylabel(r"$\rm std$", fontsize=12)
eximg = pjoin('..', '..', 'Data', 'extin3d', 'results', 'snrstd_{0}_{1}.pdf'.format(name, cloudn))
fig.savefig(eximg, dpi = 360, transparent = True)
plt.show()
```
