```python
extent = [189.5, 188.5, 2.5, 3.5]
plt.imshow(snr169,extent = extent) # 显示图片
plt.show()
```

```python
xgrid, ygrid, ar, realdis = radioar.readsnrsav('snr169')
```

```python
disbins = radioar.moddis((radioar.dismod(realdis) + 0.25)) - radioar.moddis((radioar.dismod(realdis) - 0.25))
```

```python
for i in range(len(disbins)):
    ar.dar[i] /= disbins[i]
```

```python
from SNR13.radioar import *
```

```python
adata = np.vstack(ar.dar)
madata = ma.masked_where(adata<0, adata)
allar = DataFrame(madata, index=[np.repeat(np.arange(len(ar.dar)),len(xgrid)), np.tile(ygrid, len(ar.dar))], columns=xgrid)
allar.index.names = ['cell','gb']
allar.columns.name = 'gl'
xii = (xgrid<190) & (xgrid>188) 
yii = (ygrid<4) & (ygrid>2)
zmax = np.percentile(allar.loc[np.tile(yii,16), xii], 98)
levs = zmax / 10 * np.arange(10+1)
```
