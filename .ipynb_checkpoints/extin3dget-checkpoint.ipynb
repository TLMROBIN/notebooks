```python
from dis_ext import *

def lnprob3d(para, x, y, bins):
    lp = lnprior3d(para, y)
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike3d(para, x, y, bins)

def lnlike3d(para, x, y, bins):
    ind = np.searchsorted(moddis(bins), x) - 1#所应取的消光值的位置比应插值的位置小1
    n = len(para)
    paras = np.zeros(n)
    for i in range(n):
        paras[i] = np.sum(para[0:i+1])
    model = np.take(paras, ind)
    yerr = 0.03 * model + 0.2
    return np.sum(np.log(1 / (np.sqrt(2 * np.pi) * yerr)) + (-(y - model) ** 2 / (2 * yerr ** 2)))

def lnprior3d(para, ydata):
    if np.all(para >= 0) and np.all(para < np.max(ydata)):
        return 0.0
    return -np.inf

res = readsav('e:/BNUCloud/work/snr/yubin/discalib/result/snr169.sav')

res = res['res']

ar = res.ar

dis = res.dis

rel = 0.15

dx = 0.5 * rel

bin = rel

gl = res.rd

gl = np.vstack(res.rd)[:,0]

gb = np.vstack(res.rd)[:,1]

muma = dismod(dis)

glmin = np.floor(np.min(gl))

glmax = np.floor(np.max(gl)) + 1

gbmin = np.floor(np.min(gb))

gbmax = np.floor(np.max(gb)) + 1

ngl = (glmax - glmin) / rel + 1

ngb = (gbmax - gbmin) /rel +1

xgri = glmin + np.arange(int(ngl)) * rel + dx  # centers of grids

ygri = gbmin + np.arange(int(ngb)) * rel + dx

nd = 16

ar3d = np.zeros((int(ngb),int(ngl),nd))

ar3d[:] = np.nan

dbin = 0.5

mubin =  5.0 + 0.5 * dbin + np.arange(nd) * dbin

realbin = mubin - 0.5 * dbin

realbin = np.append(realbin, mubin[-1] + 0.5 * dbin)

ndim = len(mubin)

nwalkers = 100
```

```python
fig = plt.figure()
disdata = []
ardata = []
for i in range(int(ngl)):
    gl0 = xgri[i]
    gla, glb = gl0 - dx, gl0 + dx
    glindex = np.logical_and(gl >= gla, gl < glb)
    if np.count_nonzero(glindex) > 2:
        ari = ar[glindex]
        disi = dis[glindex]
        mui = muma[glindex]
        gbi = gb[glindex]
        gli = gl[glindex]
        for j in range(int(ngb)):
            gb0 = ygri[j]
            gba, gbb = gb0 - dx, gb0 + dx
            gbindex = np.logical_and(gbi >= gba, gbi < gbb)
            if np.count_nonzero(gbindex) > 2:
                arij = ari[gbindex]
                disij = disi[gbindex]
                muij = mui[gbindex]
                gindex = np.logical_or(disij > 1, arij < 0.5 + 1.5 * disij)
                bindex = np.logical_and(muij > realbin[0], muij < realbin[-1])
                index = np.logical_and(gindex, bindex)
                if np.count_nonzero(index) >= 3:
                    disk = disij[index]
                    ark = arij[index]
                    maxguess = np.max(ark)
                    minguess = 0.
                    x = disk
                    x = x[:, np.newaxis]
                    y = ark
                    a, _, _, _ = np.linalg.lstsq(x, y)
                    p0 = a * (moddis(realbin[1:]) - moddis(realbin[0:-1]))
                    pos = [np.array(p0) + 1e-3*np.random.randn(ndim) for iii in range(nwalkers)]
                    sampler = emcee.EnsembleSampler(nwalkers, ndim, lnprob3d, args=(disk, ark, realbin))
                    sampler.run_mcmc(pos, 5000)
                    samples = sampler.chain[:, 3000:, :].reshape((-1, ndim))
                    resultss = np.percentile(samples, [16, 50, 84], axis = 0)
                    ar3d[j,i,:] = resultss[1]
                    ax = fig.add_subplot(111)
                    disar = ax.plot(disk, ark, '.r')
                    realstep = np.zeros_like(resultss[1])
                    for k in range(len(realstep)):
                        realstep[k] = np.sum(resultss[1][0:k+1])
                    ax.step(moddis(realbin)[0:len(realbin) - 1], realstep)
                    tit = '{0:7.3} {1:5.3f}'.format(gl0, gb0)
                    ax.set_title(tit)
                    xlim = (0,4)
                    ylim = (0,3)
                    plt.setp(fig.axes, xlim = xlim)
                    plt.setp(fig.axes, ylim = ylim)
                    fig.savefig('E:/BNUCloud/work/git/Data/extin3d/results/snr169/snr169_{0}_{1}.png'.format(i, j))
                    ax.cla()
```
