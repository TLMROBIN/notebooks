## 作业第一部分
1. 读取文本，以逗号分割
2. 以“李白”特征，找出每首诗，并提取诗文部分并拼接成一整段字符串
3. 用list()将所有的字符变成单独列表元素
4. 利用ASCII码提取其中的汉字，去掉标点符号
5. 利用集合的唯一性，用set()找出李白诗文中用过至少一次的汉字
6. 分别判断这些汉字是否只在其诗文中出现过一次
7. 按要求输出结果

```python
#coding=utf-8
with open('李白.txt',encoding='utf-8', mode = 'r') as f:
    text = f.read().split(',')

poemIndex = [i+1 for i,t in enumerate(text) if t == '李白']
poemText = ''.join([text[i] for i in poemIndex ])
words = list(poemText)
chineseWords = [w for w in words if '\u4e00' <= w <='\u9fff']
wordSingle = [word for word in set(chineseWords) if words.count(word) == 1]

print('诗歌部分除去标点符号的汉字共有',len(chineseWords),'字')
print('其中不重复，即只使用过一次，的汉字有',len(wordSingle),'字')
print('它们分别是：',)
for i in range(len(wordSingle)):
    print(wordSingle[i],end='  ') if i%10 != 9 else print(wordSingle[i])
```

## 作业第二部分
1. 在第一部分用.count()统计每个汉字的使用次数的基础上，统计诗歌的数量
2. 二者相除得到各个汉字在每首诗中的使用频率
3. 利用sorted()进行逆行排序，从高到低，取前50个字
4. 将得到的结果化为字典，方便一一对应
5. 按要求打印结果
6. 将上述代码改写为函数poemcount()，根据输入诗人名字的不同统计两位诗人的词频
7. 增加功能，讲词频大于0.1的词以集合的形式返回
8. 利用集合的特性，将返回的集合并、差，得到“两位诗人都爱用的字”、“李白爱用杜甫不爱用的字”、“杜甫爱用而李白不败用的字”
9. 按要求输出结果

```python
#coding=utf-8
# 第1题 在第一部分基础上增加字频统计
with open('李白.txt',encoding='utf-8', mode = 'r') as f:
    text = f.read().split(',')

poemIndex = [i+1 for i,t in enumerate(text) if t == '李白']
poemText = ''.join([text[i] for i in poemIndex ])
words = list(poemText)
chineseWords = [w for w in words if '\u4e00' <= w <='\u9fff']
wordSingle = [word for word in set(chineseWords) if words.count(word) == 1]

print('诗歌部分除去标点符号的汉字共有',len(chineseWords),'字')
print('共使用', len(set(chineseWords)),'个汉字')
print('其中不重复的汉字有',len(wordSingle),'字')
print('它们分别是：',)
for i in range(len(wordSingle)):
    print(wordSingle[i],end='  ') if i%10 != 9 else print(wordSingle[i])
# 新增字频统计部分
wordFrequency = dict([(word, words.count(word)/text.count('唐代')) for word in set(chineseWords)])
frequencySorted = dict(sorted(wordFrequency.items(), key=lambda d:d[1], reverse = True)[:50])
print('李白使用频率最高的50个字和它们在一首诗中的使用频率为：')
for key in frequencySorted.keys():
    print('{0}    {1:6.4f}'.format(key, frequencySorted[key]))
```

```python
# coding=utf-8
# 第2题
# 将第1题内容打包为函数
def poemcount(name):
    with open('{name}.txt'.format(name=name),encoding='utf-8', mode = 'r') as f:
        text = f.read().split(',')

    poemIndex = [i+1 for i,t in enumerate(text) if t == '{name}'.format(name=name)]
    poemText = ''.join([text[i] for i in poemIndex ])
    words = list(poemText)
    chineseWords = [w for w in words if '\u4e00' <= w <='\u9fff']
    wordSingle = [word for word in set(chineseWords) if words.count(word) == 1]
    wordFrequency = [(word, words.count(word)/len(chineseWords)) for word in set(chineseWords)]

    print('{name}诗歌部分除去标点符号的汉字共有'.format(name = name), len(chineseWords),'字')
    print('共使用', len(set(chineseWords)),'个汉字')
    print('其中不重复的汉字有',len(wordSingle),'字')
    print('它们分别是：',)
    for i in range(len(wordSingle)):
        print(wordSingle[i],end='  ') if i%10 != 9 else print(wordSingle[i])
    print('\n')
    wordFrequency = dict([(word, words.count(word)/text.count('唐代')) for word in set(chineseWords)])
    frequencySorted = sorted(wordFrequency.items(), key=lambda d:d[1], reverse = True)
    top50 = dict(sorted(wordFrequency.items(), key=lambda d:d[1], reverse = True)[:50])
    print('{name}使用频率最高的50个字和它们在一首诗中的使用频率为：'.format(name=name))
    for key in top50.keys():
        print('{0}    {1:6.4f}'.format(key, top50[key]))
        
    # 获得频率大于0.1的集合
    f01 = set([word for word, frequency in frequencySorted if frequency > 0.1])
    return f01

dufu = poemcount('杜甫')
libai = poemcount('李白')
```

```python
# 两位诗人都爱用的字
both = list(libai & dufu)
print('两位诗人都爱用的字:')
for i in range(len(both)):
    print(both[i],end='  ') if i%10 != 9 else print(both[i])
print('\n')
# 李白爱用杜甫不爱用的字
LILOVE = list(libai - dufu)
print('李白爱用而杜甫不爱用的字:')
for i in range(len(LILOVE)):
    print(LILOVE[i],end='  ') if i%10 != 9 else print(LILOVE[i])
print('\n')
# 杜甫爱用而李白不败用的字
DULOVE = list(dufu - libai)
print('杜甫爱用而李白不败用的字:')
for i in range(len(DULOVE)):
    print(DULOVE[i],end='  ') if i%10 != 9 else print(DULOVE[i])
print('\n')
```

## 作业第三部分
1. 将第二部分的函数略作修改，增加返回包含词频最高的50个字和它们的词频的字典
2. 将得到的字典利用wordcloud模组的WordCloud()函数创建成词墙对象
3. 调整字体解决中文乱码问题
4. 用matplotlib画出词墙

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
def poemcount(name):
    with open('{name}.txt'.format(name=name),encoding='utf-8', mode = 'r') as f:
        text = f.read().split(',')

    poemIndex = [i+1 for i,t in enumerate(text) if t == '{name}'.format(name=name)]
    poemText = ''.join([text[i] for i in poemIndex ])
    words = list(poemText)
    chineseWords = [w for w in words if '\u4e00' <= w <='\u9fff']
    wordSingle = [word for word in set(chineseWords) if words.count(word) == 1]
    wordFrequency = [(word, words.count(word)/len(chineseWords)) for word in set(chineseWords)]

    print('{name}诗歌部分除去标点符号的汉字共有'.format(name = name), len(chineseWords),'字')
    print('共使用', len(set(chineseWords)),'个汉字')
    print('其中不重复的汉字有',len(wordSingle),'字')
    print('它们分别是：',)
    for i in range(len(wordSingle)):
        print(wordSingle[i],end='  ') if i%10 != 9 else print(wordSingle[i])
    print('\n')
    wordFrequency = dict([(word, words.count(word)/text.count('唐代')) for word in set(chineseWords)])
    frequencySorted = sorted(wordFrequency.items(), key=lambda d:d[1], reverse = True)
    top50 = dict(sorted(wordFrequency.items(), key=lambda d:d[1], reverse = True)[:50])
    print('{name}使用频率最高的50个字和它们在一首诗中的使用频率为：'.format(name=name))
    for key in top50.keys():
        print('{0}    {1:6.4f}'.format(key, top50[key]))
        
    # 获得频率大于0.1的集合
    f01 = set([word for word, frequency in frequencySorted if frequency > 0.1])
    return f01,top50
```

```python
_,dufuF = poemcount('杜甫')
_,libaiF = poemcount('李白')
```

```python
wcLI = WordCloud(font_path='c:/Windows/Fonts/simsun.ttc')
wcLI.fit_words(libaiF)
plt.imshow(wcLI, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wcDU = WordCloud(font_path='c:/Windows/Fonts/simsun.ttc')
wcDU.fit_words(dufuF)
plt.imshow(wcDU, interpolation='bilinear')
plt.axis('off')
plt.show()
```
